```python
DIR = "./results/"
results_dir = "./figures/"
folders = os.listdir(DIR)


fig1, axes1 = plt.subplots(nrows=1, ncols=1, figsize=(6.6, 6.6))
fig2, axes2 = plt.subplots(nrows=1, ncols=1, figsize=(6.6, 6.6))
fig3, axes3 = plt.subplots(nrows=1, ncols=1, figsize=(6.6, 6.6))
fig4, axes4 = plt.subplots(nrows=1, ncols=1, figsize=(6.6, 6.6))

for fold in folders:
    cwd = DIR+fold+'/'
    for file in (next(os.walk(cwd))[2]):
        data = pd.read_csv(cwd+file)
        if file == "lgn_bpr.csv":
            axes1.plot(data['Epoch'].values+10,data['Loss'].values,c='r',label=file)

            axes2.plot(data['Epoch'].values+10,data['Recall'].values,c='r',label=file)

            axes3.plot(data['Epoch'].values+10,data['Precision'].values,c='r',label=file)

            axes4.plot(data['Epoch'].values+10,data['NDCG'].values,c='r',label=file)
        elif file == "mf_bpr.csv":
            axes1.plot(data['Epoch'].values,data['Loss'].values,c='r',label=file)

            axes2.plot(data['Epoch'].values,data['Recall'].values,c='r',label=file)

            axes3.plot(data['Epoch'].values,data['Precision'].values,c='r',label=file)

            axes4.plot(data['Epoch'].values,data['NDCG'].values,c='r',label=file)
        
        else :
            axes1.plot(data['Epoch'].values,data['Loss'].values,c='k',label=file)

            axes2.plot(data['Epoch'].values,data['Recall'].values,c='k',label=file)

            axes3.plot(data['Epoch'].values,data['Precision'].values,c='k',label=file)

            axes4.plot(data['Epoch'].values,data['NDCG'].values,c='k',label=file)



axes1.grid(True)
axes1.set_ylabel('Training Loss')
axes1.set_xlabel('No. of epochs')
leg1 = axes1.legend()
axes1.set_title('Plot of Training Loss against Epochs')       


axes2.grid(True)
axes2.set_ylabel('Recall')
axes2.set_xlabel('No. of epochs')
leg2 = axes2.legend()
axes2.set_title('Plot of Recall against Epochs') 


axes3.grid(True)
axes3.set_ylabel('Precision')
axes3.set_xlabel('No. of epochs')
leg3 = axes3.legend()
axes3.set_title('Plot of Precision against Epochs') 


axes4.grid(True)
axes4.set_ylabel('NDCG')
axes4.set_xlabel('No. of epochs')
leg4 = axes4.legend()
axes4.set_title('Plot of NDCG against Epochs') 
        
        
now = datetime.datetime.now()
```

```python
fig1.savefig(results_dir + '_' + str(now) + '_LOSS.pdf')
fig2.savefig(results_dir + '_' + str(now) + '_RECALL.pdf')
fig3.savefig(results_dir + '_' + str(now) + '_PRECISION.pdf')
fig4.savefig(results_dir + '_' + str(now) + '_NDCG.pdf')
```
